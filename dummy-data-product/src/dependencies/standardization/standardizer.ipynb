```python
# Heatmap graph

plt.figure(figsize=(14,8))
sns.set_theme(style="white")
corr = concat_df.corr()
heatmap = sns.heatmap(corr, annot=True, cmap="Blues", fmt='.1g')
```

```python
sns.set_theme(style="white")
corr = concat_df.corr()
mask = np.triu(concat_df.corr())
f, ax = plt.subplots(figsize=(10, 10))
cmap = sns.color_palette("Blues")

sns.heatmap(corr, 
            mask=mask, 
            cmap=cmap, 
            vmax=.3, 
            center=0,
            square=True, 
            linewidths=3, 
            cbar_kws={"shrink": .5}
           )
```
